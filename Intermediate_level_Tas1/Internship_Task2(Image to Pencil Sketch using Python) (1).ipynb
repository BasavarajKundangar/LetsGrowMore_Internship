## 1. Importing the Image

```python
image = cv2.imread("iron_man.jpg")
Image('D:\Internship\Task 1\iron_man.jpg',width=250,
    height=50)
```

```python
#image width and height
print("Width of image:",image.shape[1])
print("Height of image:",image.shape[0])
```

## 2. Adding grey filter

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
data = im.fromarray(grey_image).resize(size=(250,400))
data
```

## 3 . Adding inverted filter

```python
inverted_image = 250 - grey_image
data = im.fromarray(inverted_image).resize(size=(250,400))
data
```

## 4.Adding Blur effect

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred_image = 255 - blurred
data = im.fromarray(inverted_blurred_image).resize(size=(250,400))
data
```

# Final Pencil Sketch

```python
pencil_sketch = cv2.divide(grey_image, inverted_blurred_image, scale=256.0)
data = im.fromarray(pencil_sketch).resize(size=(250,400))
data
```

## 5. Exporting the Image

```python
# Saving the image
cv2.imwrite('Final_pencil_sketch.png',pencil_sketch)
```
